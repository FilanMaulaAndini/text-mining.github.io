```python
import pandas as pd 
import numpy as np
import string as str
import re
import Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#parsing dilakukan pada saat pengumpulan dokumen di file x.csv

# method untuk melakukan tokenisasi dengan menggunakan split
def tokenize(doc) : 
    doc = doc.split()
    return doc

#method untuk menghilangkan tanda baca, karakter, angka, dan menghilangkan huruf pada kata lebih dari 1
def cleansing(doc):
    doc = re.sub(r'[^a-z]+', ' ', doc)
    doc = re.sub(r'([a-z])\1+',r'\1',doc)
    return doc

#method untuk mengubah semua huruf menjadi huruf kecil
def case_folding(doc):
    doc = doc.lower()
    return doc

#method untuk menghilangkan kata yang tidak penting, stoplist diambil dari web http://static.hikaruyuuki.com/wp-content/uploads/stopword_list_tala.txt/
def filtering(doc):
    stop = open('daftarstoplist.txt','r')
    sw = stop.read()
    stop.close()

    sw = sw.split("\n")
    filtering = []
    for i in doc:
        if i not in sw:
            filtering.append(i)
    return filtering

#method untuk mengubah kata yang berimbuhan menjadi kata dasar menggunakan library sastrawi
def stemming(doc):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    katadasar = []
    for data in doc:
        kd = stemmer.stem(data)
        katadasar.append(kd)
    return katadasar

#method untuk menghilangkan kata yang lebih dari 1
#def remove_duplicate(doc):
    #term = []
    #for data in doc:
        #if(data not in term):
            #term.append(data)
    #return term


#method untuk menjalankan proses preoprocessing
def preprocessing(doc):
    case = case_folding(doc)
    clean = cleansing(case)
    token = tokenize(clean)
    stop = filtering(token)
    stem = stemming(stop)
    return stem

#melakukan looping pada file c.csv untuk menjalankan method preprocessing
new_doc = pd.read_csv('x.csv', header=None, sep="\n")
for i in range(len(new_doc)):
    print("> Hasil dari preprocessing dokumen ke-",i+1,':', preprocessing(new_doc[0][i]))
    print("------------------------------------------------------------------------------------------------------------")
```

```python
#digunakan untuk menggabungkan seluruh term pada dokumen dalam 1 array
doc = []
for i in range(len(new_doc[0])):
    doc.append(new_doc[0][i])
    doc[i] = preprocessing(doc[i])
```

```python
#method untuk menghitung raw frequency term
def raw_frequency(doc, result, kolum): 
    for i in range(len(doc)):
        for j in range(len(doc[i])):
            for k in range(len(result)):
                if(doc[i][j]==result.iloc[k][0]):
                    result.at[k , kolum[i]] = result.iloc[k][i+1] + 1
    return result
```

```python
#digunakan untuk menampilkan method raw frequency term 
raw_frequency(doc, result, kolum)
result.to_csv('raw.csv')
#result.iloc[1][1]
```

```python
#method untuk menghitung log frequency term 
kolum = ['D1', 'D2','D3','D4','D5','D6','D7','D8','D9','D10','D11','D12','D13','D14','D15','D16','D17','D18','D19','D20','D21','D22','D23','D24','D25','D26','D27','D28','D29','D30']
def log_frequency(log, kolum):
    for i in range(len(log)):
        for j in range(len(kolum)):
            if(log.iloc[i][j+1] != 0):
                log.at[i , kolum[j]] = 1 + math.log10(log.iloc[i][j+1])
    return log
```

```python
#digunakan untuk menampilkan log term frequency
log_frequency(log, kolum)
```

```python
#method untuk menghitung df
# kolum = ['D1', 'D2','D3','D4','D5','D6','D7','D8','D9','D10','D11','D12','D13','D14','D15','D16','D17','D18','D19','D20','D21','D22','D23','D24','D25','D26','D27','D28','D29','D30']
def count_df(log, kolum):
    df = []
    count = 0
    for i in range(len(log)):
        for j in range(len(kolum)):
            if(log.iloc[i][j+1] != 0):
                count += 1
        df.append(count)
        count = 0
    return df
            
#method untuk menghitung idf
def count_idf(doc, df):
    idf = []
    df=count_df(log,kolum)
    for i in range(len(df)):
        x=math.log10(len(doc)/df[i])
        idf.append(x)
    return idf
```

```python
def pisahDokumen(doc):  
    doc_uji = []
    for i in range(len(doc[0])):
        doc_uji.append(doc[0][i])
        doc_uji[i] = preprocessing(doc_uji[i])
    return doc_uji
```

```python
def pisahTerm(doc):
    term_list_uji = []
    doc_uji=pisahDokumen(doc)
    for i in range(len(doc_uji)):
        for j in range(len(doc_uji[i])):
            term_list_uji.append(doc_uji[i][j])
    return term_list_uji
```

```python
def labelUji(hasilUji):
    label = []
    count = 0
    for i in range(len(hasilUji)):
            if(hasilUji.iloc[i][0]):
                count = max(hasilUji.iloc[i][0],hasilUji.iloc[i+1][0],hasilUji.iloc[i+2][0])
                if (count == hasilUji.iloc[i][0]):
                    count = kolumX[0]
                elif (count == hasilUji.iloc[i+1][0]):
                    count = kolumX[1]
                else:
                    count = kolumX[2]
            label.append(count)
            count = 0
            return label
```
