```python
import pandas as pd 
import numpy as np
import string as str
import re
import Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#parsing dilakukan pada saat pengumpulan dokumen di file data_latih.csv

# method untuk melakukan tokenisasi dengan menggunakan split
def tokenize(doc) : 
    doc = doc.split()
    return doc

#method untuk menghilangkan tanda baca, karakter, angka, dan menghilangkan huruf pada kata lebih dari 1
#contoh menghilangkan huruf lebih dari 1 pada suatu term adalah ('pagiiii' menjadi 'pagi' )
def cleansing(doc):
    doc = re.sub(r'[^a-z]+', ' ', doc)
    doc = re.sub(r'([a-z])\1+',r'\1',doc)
    return doc

#method untuk mengubah semua huruf menjadi huruf kecil
def case_folding(doc):
    doc = doc.lower()
    return doc

#method untuk menghilangkan kata yang tidak penting, stoplist diambil dari web https://www.kaggle.com/oswinrh/indonesian-stoplist
def filtering(doc):
    stop = open('daftarstoplist.txt','r')
    sw = stop.read()
    stop.close()

    sw = sw.split("\n")
    filtering = []
    for i in doc:
        if i not in sw:
            filtering.append(i)
    return filtering

#method untuk mengubah kata yang berimbuhan menjadi kata dasar menggunakan library sastrawi
def stemming(doc):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    katadasar = []
    for data in doc:
        kd = stemmer.stem(data)
        katadasar.append(kd)
    return katadasar

#method untuk menjalankan proses preoprocessing
def preprocessing(doc):
    case = case_folding(doc)
    clean = cleansing(case)
    token = tokenize(clean)
    stop = filtering(token)
    stem = stemming(stop)
    return stem

#melakukan looping pada file c.csv untuk menjalankan method preprocessing
new_doc = pd.read_csv('data_latih.csv', header=None, sep="\n")
for i in range(len(new_doc)):
    print("> Hasil dari preprocessing dokumen ke-",i+1,':', preprocessing(new_doc[0][i]))
    print("------------------------------------------------------------------------------------------------------------")
```

```python
#digunakan untuk menggabungkan seluruh term pada dokumen dalam 1 array
doc = []
for i in range(len(new_doc[0])):
    doc.append(new_doc[0][i])
    doc[i] = preprocessing(doc[i])
```

```python
#method untuk menghitung raw frequency term
def raw_frequency(doc, result, kolum): 
    for i in range(len(doc)):
        for j in range(len(doc[i])):
            for k in range(len(result)):
                if(doc[i][j]==result.iloc[k][0]):
                    result.at[k , kolum[i]] = result.iloc[k][i+1] + 1
    return result
```

```python
#menghitung posterior term yang digunakan pada data uji untuk kategori Makanan
def probabilityUjiMakanan(term_list_uji):
    nilai=1
    for i in range(len(term_list_uji)):
        for j in range(len(datalatih)):
            if(term_list_uji[i]==datalatih.iloc[j][0]):
                nilai*=datalatih.iloc[j][1]
    return nilai
```

```python
#menghitung posterior term yang digunakan pada data uji untuk kategori Olahraga
def probabilityUjiOlahraga(term_list_uji):
    nilai=1
    for i in range(len(term_list_uji)):
        for j in range(len(datalatih)):
            if(term_list_uji[i]==datalatih.iloc[j][0]):
                nilai*=datalatih.iloc[j][2]
    return nilai
```

```python
#menghitung posterior term yang digunakan pada data uji untuk kategori Perjalanan
def probabilityUjiPerjalanan(term_list_uji):
    nilai=1
    for i in range(len(term_list_uji)):
        for j in range(len(datalatih)):
            if(term_list_uji[i]==datalatih.iloc[j][0]):
                nilai*=datalatih.iloc[j][3]
    return nilai
```

```python
#method untuk memberi label pada data uji sesuai dengan nilai posterior terbesar
def labelUji(hasilUji):
    label = []
    count = 0
    for i in range(len(hasilUji)):
            if(hasilUji.iloc[i][0]):
                count = max(hasilUji.iloc[i][0],hasilUji.iloc[i+1][0],hasilUji.iloc[i+2][0])
                if (count == hasilUji.iloc[i][0]):
                    count = kolumX[0]
                elif (count == hasilUji.iloc[i+1][0]):
                    count = kolumX[1]
                else:
                    count = kolumX[2]
            label.append(count)
            count = 0
            return label
```

```python
#method untuk menjalankan proses data uji 
def testUji(doc):
    pre = preprocessing(doc)
    probMa = probabilityUjiMakanan(pre) * (len(kolumA)/len(kolum))
    probOl = probabilityUjiOlahraga(pre) * (len(kolumB)/len(kolum))
    probPer = probabilityUjiPerjalanan(pre) * (len(kolumC)/len(kolum))
    hasilUji = pd.DataFrame([probMa,probOl,probPer])
    label = labelUji(hasilUji)
    return label
```

```python
#membaca file .csv yagn berisi doumen uji
datauji = pd.read_csv('data_uji.csv', header=None)
#method untuk memberi label pada dokumen uji dengan melakukan perulangan untuk tiap dokumennya
def beriLabelUji(datauji):
    labelBaru=[]
    for i in range(len(datauji)):
        x=testUji(datauji[0][i])
        labelBaru.append(x)
    return labelBaru
```
